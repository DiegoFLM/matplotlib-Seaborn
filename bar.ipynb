```python
import numpy as np
import matplotlib.pyplot as plt
```

# Bar Plot

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1410, 126, 132, 52, 84]
```

```python
plt.style.use('bmh')
```

```python
barGraph = plt.bar(country, population, width = 0.5, 
    color = ['red', 'blue', 'green', 'yellow', 'purple'])
plt.xticks(np.arange(5), ('India', 'Japan', 'Mexico', 'Colombia', 'Germany'), rotation = 45) # x axis config
plt.title('Population of five countries')
plt.xlabel('Country')
plt.ylabel('Population [millions people]')
plt.bar_label(barGraph, labels = population)
plt.show()
```
