```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips.dtypes
```

```python
tips.describe()
```

# sns.hisplot()

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative = True)
plt.show()
```

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative = False, hue = 'sex', 
stat = 'probability', multiple = 'layer'
)
plt.show()
```

# sns.kdeplot()

```python
sns.kdeplot(data = tips, x = 'tip', hue = 'sex', legend = True, cumulative = False,
shade = True, bw_adjust = 1
)
plt.show()
```

# sns.ecdfplot()

```python
sns.ecdfplot(data = tips, x = 'tip', hue = 'sex', stat = 'count')
plt.show()
```

# sns.displot()

```python
sns.displot(data = tips, x = 'tip', hue = 'sex', kind = 'kde', multiple = 'stack')
plt.show()
```

# categorical graphics

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
#sns.countplot(data = tips, y = 'day', hue = 'sex')
sns.countplot(data = tips, x = 'day', hue = 'sex')
plt.show()
```

```python
plt.figure(figsize = (6,6))
sns.stripplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex', alpha = 0.6, dodge = True)
plt.show()
```

```python
plt.figure(figsize = (6,6))
sns.swarmplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex', alpha = 1, dodge = True)
plt.show()
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex',  dodge = True)
plt.show()
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex',  dodge = True)
sns.swarmplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex', alpha = 1, 
    dodge = True, color = '0')
plt.show()
```

```python
plt.figure(figsize = (6,6))
sns.violinplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex',  dodge = True, 
palette = 'pastel', split = True)
plt.show()
```

```python
sns.catplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex', dodge = True, #alpha = 0.6,
kind = 'box', col = 'time'
)
plt.show()
```
