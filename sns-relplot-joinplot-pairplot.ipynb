# sns.scatterplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips[tips['time'] == 'Lunch']
```

```python
plt.figure(figsize = (15,9))
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', 
style = 'time', alpha = 0.7, size = 'size') #size refers to marker size. 'size' is the amount of people.
#plt.legend(loc = 'center', bbox_to_anchor = (1.12, 0.5))
plt.show()
```

```python
markers = {'Lunch': 'D', 'Dinner':'s'}
plt.figure(figsize = (15,9))
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', 
style = 'time', alpha = 0.7, size = 'size', markers = markers
) #size refers to marker size. 'size' is the amount of people.
plt.show()
```

# sns.relplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python

sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', 
style = 'time', alpha = 0.7, size = 'size', height = 5, aspect = 1.3, kind = 'line', col = 'time')
#sns.set(rc={"figure.figsize":(15, 9)})
#plt.figure(figsize = (15, 9))
plt.show()
```

# joinplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', marginal_ticks = True)
plt.show()
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', marginal_ticks = True,
    marginal_kws = dict( bins = 25, fill = False, multiple = 'dodge' )
)
plt.show()
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex')
plt.show()
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', kind = 'hist')
plt.show()
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', kind = 'kde')
plt.show()
```

# pairplot

pairplot works with numeric variables

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.pairplot(data = tips)
plt.show()
```

```python
sns.pairplot(data = tips, hue = 'sex')
plt.show()
```

```python
sns.pairplot(data = tips, hue = 'sex', palette = 'coolwarm', corner = True,
kind = 'scatter'
)
plt.show()
```
