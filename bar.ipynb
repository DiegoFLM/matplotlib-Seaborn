```python
import numpy as np
import matplotlib.pyplot as plt
```

# Bar Plot

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1410, 126, 132, 52, 84]
```

```python
plt.style.use('bmh')
```

```python
barGraph = plt.bar(country, population, width = 0.5, 
    color = ['red', 'blue', 'green', 'yellow', 'purple'])
plt.xticks(np.arange(5), ('India', 'Japan', 'Mexico', 'Colombia', 'Germany'), 
    rotation = 45) # x axis config
plt.title('Population of five countries')
plt.xlabel('Country')
plt.ylabel('Population [millions people]')
plt.bar_label(barGraph, labels = population)
plt.show()
```

```python
plt.barh(country, population)
```

## Histogram

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randint(1, 51, 100)
data.max()
```

```python
data
```

```python
plt.hist(data, bins = 10, width = 4)
plt.show()
```

## boxplot

```python
copDat = data
copDat = np.append(copDat, 200)
copDat
```

```python

plt.boxplot(copDat, vert = False, patch_artist = True, notch = True)
plt.show()
```

```python
plt.boxplot(copDat, vert = False, patch_artist = True, notch = True, showfliers = False)
plt.show()
```

## scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2
colors = np.random.rand(N)
```

```python
a = x[:20]
b = y[:20]
area1 = area[:20]
c = x[20:]
d = y[20:]
area2 = area[20:]

plt.scatter(a, b, s = area1, c = 'purple', alpha = 0.6)
plt.scatter(c, d, s = area2, c = 'red', alpha = 0.5)

plt.show()
```
