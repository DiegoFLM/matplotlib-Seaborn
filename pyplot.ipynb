```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
x
```

```python
plt.plot(x, y)
plt.show()
```

```python
#The third argument is composed; 'b' makes the graphic to be blue, and 'o' makes it
#be shown in circles, and '-' shows a line.
plt.plot(x, y, 'bo-')
plt.show()
```

```python
plt.plot(x, y, 'yD:') #'D' := Diamonds
plt.show()
```

## plt.scatter

```python
plt.scatter(x, y)
plt.show()
```

# Histogram, Pie, Boxplot

# Subplot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
y
```

```python
#First arg:     Number of rows
#Second arg:    Number of columns
#Third arg:     Index of the current element
plt.subplot(1, 2, 1)
plt.plot(x, y, 'b--')
plt.subplot(1, 2, 2)
plt.hist(y)
plt.show()
```

```python
y
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'b--')
plt.plot(y, x, 'r:')
plt.subplot(1, 2, 2)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, y, 'b--')
plt.plot(y, x, 'r:')
plt.subplot(2, 1, 2)
plt.pie(y)
plt.show()
```
