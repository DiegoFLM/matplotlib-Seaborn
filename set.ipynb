```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips.dtypes
```

## sns.displot()

```python
sns.displot(data = tips, x = 'total_bill')
plt.show()
```

```python
sns.displot(data = tips, x = 'total_bill', y = 'tip')
plt.show()
```

```python
sns.displot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex')
plt.show()
```

```python
sns.displot(data = tips, x = 'total_bill', hue = 'sex')
plt.show()
```

```python
sns.displot(data = tips, x = 'total_bill', hue = 'sex', kind = 'kde')
plt.show()
```

```python
sns.displot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', kind = 'kde', legend = True, 
    palette = 'dark', alpha = 0.9)
plt.show()
```
