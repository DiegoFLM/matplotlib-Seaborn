```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
x
```

```python
plt.plot(x, y)
plt.show()
```

```python
#The third argument is composed; 'b' makes the graphic to be blue, and 'o' makes it
#be shown in circles, and '-' shows a line.
plt.plot(x, y, 'bo-')
plt.show()
```

```python
plt.plot(x, y, 'yD:') #'D' := Diamonds
plt.show()
```

## plt.scatter

```python
plt.scatter(x, y)
plt.show()
```

# Histogram, Pie, Boxplot

```python
print(x)
plt.hist(x)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```
