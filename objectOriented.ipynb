```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 11)
y = x**2
y
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.5, 1])
axes.plot(x, y, 'b')
fig.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.9])
axes2 = fig.add_axes([0.2, 0.55, 0.4, 0.3])

axes.plot(x, y, 'b')
axes2.plot(y, x, 'r')
axes2.set_facecolor('gray')

fig.show()
```

# subplots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
y
```

```python
fig, axes = plt.subplots()
axes.plot(x, y, 'b')
fig.show()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x, y, 'b')
axes[1].plot(y, x, 'r')
```

```python
fig, axes = plt.subplots(2,4)
axes[0,0].plot(x, np.cos(x))
axes[0,1].plot(x, np.sin(x), 'y')
axes[0,2].plot(x, np.tan(x), 'go')
axes[0,3].plot(x, np.cos(x)**2, "ro")
axes[1,0].plot(x, np.log(x))
axes[1,1].plot(x, np.e ** np.sin(x), 'y')
axes[1,2].plot(x, np.tan(x) ** 0.5, 'go')
axes[1,3].plot(x, np.cos(x**2), "ro")
fig.tight_layout()
fig.show()
```
