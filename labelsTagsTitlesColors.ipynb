```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
y
```

```python
fig, axes = plt.subplots(1, 2, figsize = (10,5))
axes[0].plot(x, y, label = "$sin(x)$") #the label needs the legend() function to be seen
axes[0].set_title('sin(x) vs x')
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')
axes[0].legend()

axes[1].plot(y, x)
axes[1].set_title('x vs sin(x)')
axes[1].set_xlabel('sin(x)')
axes[1].set_ylabel('x')
```

```python
plt.plot(x, y, label = 'sin(x)')    #the label needs plt.legend() to be seen
plt.title('sin(x) vs x')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()#loc = 'lower left', bbox_to_anchor = (1, 0.2))
```

# Colors & Styles

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
y
```

```python
print(plt.style.available)
plt.style.use('bmh')
```

```python
for i in plt.style.available:
  plt.style.use(i)
  fig, ax = plt.subplots(figsize=(5,5))
  ax.plot(x, y)
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)

for s in plt.style.available:
  plt.style.use(s)
  fig, ax = plt.subplots(figsize=(2,2))
  ax.plot(x, y)
  ax.set_title(s, size=20)
```

# Matlab

```python
plt.style.use('dark_background')
```

```python
fig, ax = plt.subplots(figsize = (8,8))
ax.plot(x, x + 1, 'r--')
ax.plot(x, x + 2, 'b-')
ax.plot(x, x + 3, 'y.-')
ax.plot(x, x + 4, 'go:')
```

## Pyplot

```python
fig, ax = plt.subplots(figsize = (8,8))
ax.plot(x, x + 1, color = 'green', alpha = 0.8) # alpha := inverse measure of transparency
ax.plot(x, x + 2, color = 'blue', linewidth = 8, linestyle = '--', 
    marker = 'p', markersize = 12, markerfacecolor = '#CCFFCC')
ax.plot(x, x + 3, color = '#66FF66', linewidth = 3, linestyle = 'dashed', marker = 'x')    # Color can be specified as RGB
ax.plot(x, x + 4, color = '#CCFFCC', linewidth = 2, linestyle = ':', marker = 'v')
```
